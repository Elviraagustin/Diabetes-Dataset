## **Sumber Data**

Dataset yang digunakan dalam analisis ini adalah " Diabetes Prediction Dataset" yang diperoleh dari Kaggle. Dataset ini berisi informasi mengenai banyaknya yang menderita penyakit diabetes, riwayat perokok, hipertensi, dan hal lainnya.

## **Struktur Data**

Dataset ini berisi informasi tentang pasien diabetes dan memiliki beberapa kolom. Berikut adalah struktur data dari dataset tersebut:

- Pregnancies: Jumlah kehamilan yang dialami pasien.

- Glucose: Kadar glukosa dalam darah (mg/dL) pada saat pemeriksaan.
- smoking_history: Riwayat Perokok.
- BloodPressure: Tekanan darah diastolik (mm Hg).

- hypertension: Yang mengalami Hipertensi.

- Insulin: Kadar insulin dalam darah (mu U/ml).

- BMI: Indeks massa tubuh (kg/mÂ²).

- DiabetesPedigreeFunction: Skor yang menunjukkan riwayat diabetes dalam keluarga.

- Age: Usia pasien (tahun).

- Outcome: Status diabetes (0 = tidak diabetes, 1 = diabetes).

## **Melakukan imputasi**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(df)) * 100
    results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

print(results_df)
```
